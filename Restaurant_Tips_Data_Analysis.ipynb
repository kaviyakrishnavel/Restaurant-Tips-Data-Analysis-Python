```python
import pandas as pd
import matplotlib.pyplot as plt

print("Libraries imported successfully")
```

```python
url = "http://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df =  pd.read_csv(url)

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
#Average tip
df['tip'].mean()
```

```python
#Maximum bill amount
df['total_bill'].max()
```

```python
#Minimum bill amount
df['total_bill'].min()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['total_bill'])
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
#Average tip by day
df.groupby('day')['tip'].mean()
```

```python
#Average tip by time (Lunch/Dinner)
df.groupby('time')['tip'].mean()
```

```python
#Average tip by gender
df.groupby('sex')['tip'].mean()
```

```python
#Average tip by smoker
 df.groupby('smoker')['tip'].mean()
```

```python
import matplotlib.pyplot as plt

df.groupby('day')['total_bill'].mean().plot(kind='bar')
plt.title('Average Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Average Total Bill')
plt.show()
```

```python
df[['total_bill', 'tip', 'size']].corr()
```

*Dinner time generates higher bills and tips
*Larger group size leads to higher total bill
*Tips are positively correlated with total bill

#Conclusion
This analysis explored restaurant billing and tipping behaviour.
This results show that higher total bills are associated with higher tips.
Group size also impacts total bill value.
Dinner time generally generates higher revenue compared to lunch.
